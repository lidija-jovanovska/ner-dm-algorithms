```python
# !pip install nltk
from nltk.corpus import stopwords
```

```python
vocabulary = []
with open('../../src/ner/data/ml_sample_50_annotations.txt') as f:
    for line in f.readlines():
        print(line.strip().split('\t'))
        num_tokens = int(line.strip().split('\t')[0])
        tokens = line.strip().split('\t')[1:num_tokens]
        vocabulary.append(tokens)
vocabulary = sum(vocabulary, [])
vocabulary
```

```python
len(vocabulary)
```

```python
vocabulary = [token.lower() for token in vocabulary]
```

```python
len(set(vocabulary))
```

```python
unigrams = sum([token.split(' ') for token in vocabulary], [])
unigrams.remove('')
```

```python
len(unigrams)
```

```python
filtered_unigrams = [word for word in unigrams if word not in stopwords.words('english')]
len(filtered_unigrams)
```
