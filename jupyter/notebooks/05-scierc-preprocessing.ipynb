```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
data_path = '../data/scierc/'
data_path
# train_path = os.path.join(os.getcwd(), '/data/scierc/', 'train.json')
```

Create mapping

```python
mapping = {'Task': 0,
           'Method': 1,
           'Metric': 2,
           'Material': 3,
           'OtherScientificTerm': 4,
           'Generic': 5}
```

Get names of all ann_files

```python
ann_files_paths = []
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file.endswith(".ann"):
            ann_files_paths.append(os.path.join(root, file))
```

```python
len(ann_files_paths)
```

Create dict where keys are file names and values are dicts containing the tokens and ner_tags in the file

```python
processed_data = dict.fromkeys(ann_files_paths)
```

```python
for path in ann_files_paths:
    with open(path, 'r') as f:
        ner_tags = []
        tokens = []
        for line in f.readlines():
            terms = line.split('\t')
            if terms[0].startswith('T'):
                ner_tag = terms[1].split(' ')[0]
                token = terms[2].strip()
                ner_tags.append(ner_tag)
                tokens.append(token)
#                 print(ner_tag)
#                 print(token)
        # map ner_tags
        mapped_ner_tags = [mapping[tag] for tag in ner_tags] 
        # add tokens and tags in file dict
        processed_data[path] = {'ner_tags': mapped_ner_tags, 'tokens': tokens}
#         print(processed_data[path])
```

Split train and validation

```python
train, test = train_test_split(list(processed_data.keys()), test_size=0.1, train_size=0.9, random_state=10)
```

```python
train
```

```python
def export_to_file(export_file_path, data):
    with open(export_file_path, "w") as f:
        for record in data:
            ner_tags = record["ner_tags"]
            tokens = record["tokens"]
            if len(tokens) > 0:
                f.write(
                    str(len(tokens))
                    + "\t"
                    + "\t".join(tokens)
                    + "\t"
                    + "\t".join(map(str, ner_tags))
                    + "\n"
                )
```

```python
train_data = [processed_data[x] for x in train]
train_path = os.path.join('../data/processed/', 'scierc_train.txt')
test_data = [processed_data[x] for x in test]
test_path = os.path.join('../data/processed/', 'scierc_test.txt')
```

```python
export_to_file(train_path, train_data)
```

```python
export_to_file(test_path, test_data)
```
